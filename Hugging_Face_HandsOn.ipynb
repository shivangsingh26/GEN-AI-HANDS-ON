# Example usage of table question answering from huggingface hub

```python
from transformers import pipeline
import pandas as pd

# prepare table + question
data = {"Actors": ["Brad Pitt", "Leonardo Di Caprio", "George Clooney"], "Number of movies": ["87", "53", "69"]}
table = pd.DataFrame.from_dict(data)
question = "how many movies does Leonardo Di Caprio have?"

# pipeline model
# Note: you must to install torch-scatter first.
tqa = pipeline(task="table-question-answering", model="google/tapas-large-finetuned-wtq")

# result

print(tqa(table=table, query=question)['cells'][0])
#53
```

# What is AutoTokenizer

It automatically chooses the appropriate tokenizer for the whichever model is given

```python
from transformers import AutoTokenizer
model = 'distilbert-base-uncased'
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
text = "My favourie food is Biriyani"
encoded_text = tokenizer(text)
print(encoded_text)
```

Here we have 8 tokens created

101 here is a special token ID called [CLS] which stands for classificaion and is added at  beginning of input sequence.

102 here is a special token ID called [SEP] and is used to seperate between sentences.

Here attention mask is all 1's hence we can say that all id's are given attention

**Lets see an example where attention_mask can contain zeroes ie padding is done there.**

Here instead of tokenizing a single text input we will perform tokenization on the entire dataset

```python
from datasets import load_dataset
emotion_dataset = load_dataset('emotion')
emotion_dataset
```

```python
emotion_dataset['train'][:2]
```

```python
def tokenization(batch):
  return tokenizer(batch['text'], padding = True, truncation = True)

tokenization(emotion_dataset['train'][:2])
```

here 0's in the attention_mask denotes padding.

Whichever sentence is small there padding is required to match the length with the bigger sentence.

```python
emotions_encoded = emotion_dataset.map(tokenization, batched = True, batch_size = None)
emotions_encoded
```
